## Overlay Intersect Layers and Assign Weights

```python
for k in data.keys():
    if data[k]['type'] != 'study_grid':
        layer = data[k]['geodataframe']
        
        intersect = gpd.overlay(
            grid, # grid layer
            layer, # looped layer(s)
            how='intersection'
        )
        
        data[k]['intersect_grid'] = intersect
        
        intersect.to_file(f'data/processing/grid_int_{k.lower()}_test.shp')
```

## Create new column of Intersected Area for Intersection Layer

```python
layer_list = []  # empty list 

for k, v in data.items():
    if data[k]['type'] != 'study_grid': 
        k_name = k.lower()
        layer_list.append(k_name)
        
        gdf = data[k]['intersect_grid']
        gdf[f'{k_name}_area_sqkm'] = gdf.geometry.area
        
        gdf[f'{k_name}_weighted_area'] = data[k]["weight"] * gdf[f'{k_name}_area_sqkm'] 
        
        df = gdf[[
            'uid',
            f'{k_name}_area_sqkm',
            f'{k_name}_weighted_area'
        ]].groupby(
            ['uid'],
            as_index=False,
        ).sum()
        
        data[k]['summary_weights'] = df
```

## Join Weighted Area metrics to Grid

```python
layer_list = []

for k in data.keys():
    if data[k]['type'] != 'study_grid':  # for all layers not the grid
        k_name = k.lower()
        layer_list.append(k_name)

        df = data[k]['summary_weights']
        
        grid = grid.merge(
            df[['uid', f'{k_name}_weighted_area']],
            on='uid', 
            how='left',
        )
```

#### Let's reset an initial project dictionary

```python
data_for_func = {
    "Edmonton_Grid": {
        "type": "study_grid",
    },
    "POI_Edmonton": {
        "type": "point", 
        "buffer_dist": 30.48,  # 100 ft, in meters
        "weight": 100
    },
    "Edmonton_Traffic": {
        "type": "point",   
        "buffer_dist": 30.48,  # 100 ft, in meters
        "weight": 300
    },
    "Zoning_UTM": {
        "type": "polygon",
        "weight": 2000
    }
}
```

## Let's wrap this whole process up in one function

```python
def weighted_grid_score_overlay(data):
    layer_list = []  # list for layers to sum at end for sum_weight
    
    for k in data.keys():
        data[k]['geodataframe'] = gpd.read_file(f'data/input/{k}.shp')         
         
        if data[k]['type'] == 'study_grid':
            grid = data[k]['geodataframe']
            grid['uid'] = grid.index + 1

    for k in data.keys():
        if data[k]['type'] == 'point':  # buffer point layers
            gdf = data[k]["geodataframe"]
            gdf['geometry'] = gdf.buffer(data[k]["buffer_dist"])
            data[k]['geodataframe'] = gdf
            gdf.to_file(f'data/processing/buffer_{k.lower()}.shp')

        if data[k]['type'] != 'study_grid':  # for all layers not the grid
            layer = data[k]['geodataframe']
            k_name = k.lower()  # lowercase layer name
            layer_list.append(k_name)
        
            intersect = gpd.overlay(grid, layer, how='intersection')
            intersect[f'{k_name}_area_sqkm'] = intersect.geometry.area  # calculate area
            intersect[f'{k_name}_weighted_area'] = intersect[f'{k_name}_area_sqkm'] * data[k]["weight"] 

            intersect.to_file(f'data/processing/grid_int_{k.lower()}.shp')

            df = intersect[
                ['uid', f'{k_name}_area_sqkm', f'{k_name}_weighted_area']
            ].groupby(
                ['uid'],
                as_index=False,
            ).sum()  # create summary weights dataframe
            
            data[k]['summary_weights'] = df  
            grid = grid.merge(
                df[['uid', f'{k_name}_weighted_area']],
                on='uid', 
                how='left',
            )
        
    grid['sum_weight'] = grid[[f'{i}_weighted_area' for i in layer_list]].sum(axis=1)
        
    return grid, data
```

#### Let's run this function and collect the returned `grid` and `data` objects

```python
grid_out, data_out = weighted_grid_score_overlay(data_for_func)
```
